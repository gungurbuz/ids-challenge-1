Um am Leaderboard teilzunehmen, müsst Ihr nur Eure beste Lösung auf unseren Server hochladen (unter Angabe von Gruppennamen und Zugriffs-Token).

Gruppennamen und Zugriffstoken müssen **genau wie ausgegeben** eingetragen werden, damit die Ergebnisse vom Server akzeptiert werden!

Gib Gruppennamen und Token sowie das Ergebnis Deiner Optimierung zu Beginn der Zelle ein.

Überprüfe die Ausgabe, um zu sehen, ob deine Einreichung erfolgreich war.
Falls ja, dann erscheint der Name eurer Gruppe auch direkt im Leaderboard :)

*Hinweis:* Möglicherweise müsst Ihr zunächst das Paket `requests` installieren.

Deadline für den Upload von Ergebnissen ins Leaderboard ist am Sonntag, 26.07.2026.

```python
import requests

GROUP_NAME = "YourGroupName"
GROUP_TOKEN = "YourGroupToken"  


# Dictionary containting only the three tours of the robots
sol = {
    "group_name": GROUP_NAME,
    "tours": [[0, 1, 0], [0, 2, 0], [0, 8, 4, 0]],
}


headers = {"Authorization": f"Bearer {GROUP_TOKEN}"}
response = requests.post("https://ids-challenge.imi-services.imi.kit.edu/tp01/evaluate", json=sol, headers=headers)
response.raise_for_status()
response.json()
```
