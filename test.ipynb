```python
import pandas as pd
```

```python
lsm = pd.read_csv("data/ls_matrix_55_42.csv", sep='; ', engine='python', header=0)
print(lsm)
points_ben = pd.read_csv("data/allpoint.csv", sep='; ', engine='python', header=0)
print(points_ben)
```

```python
def calc_total_tour_time(sequ, matrix):
    total_time = 0
    t = sequ.itertuples()
    for row1, row2 in zip(
        sequ.itertuples(index=0),
        sequ.iloc[1:].itertuples(index=0)
    ):
        total_time += matrix.loc[
            (matrix["start"] == row1.machine) &
            (matrix["end"] == row2.machine),
            "ls"
        ].iloc[0]
    return float(total_time)
```
